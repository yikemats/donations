```python
import requests
zuck_foundation = requests.get("https://projects.propublica.org/nonprofits/download-xml?object_id=202233199349108343")
zuck_foundation
```

```python
# The text is the raw xml of Form 990 tax document,obtained from ProPublica's non-profit explorer.
zuck_xml = zuck_foundation.text
print(zuck_xml)
```

```python
from bs4 import BeautifulSoup
```

```python
zuck_data = BeautifulSoup(zuck_xml, "xml")
type(zuck_data)
```

```python
zuck_officer_data = zuck_data.select("OfficerDirTrstKeyEmplGrp")
zuck_officer_data
```

```python
import pandas as pd
```

```python
officer_data = zuck_data.find_all("OfficerDirTrstKeyEmplGrp")

officer_list = []

for officer in officer_data:
    officer_dict = {}
    name = officer.find("PersonNm")
    officer_dict["name"] = name.text
    
    title = officer.find("TitleTxt")
    officer_dict["title"] = title.text
    
    officer_list.append(officer_dict)
```

```python
officer_list
```

```python
zuck_officer_df = pd.DataFrame(officer_list)
zuck_officer_df.to_csv('chan_zuckerberg_initiative_foundation2021_officer.csv')
zuck_officer_df
```

```python
employee_data = zuck_data.find_all("CompensationHighestPaidEmplGrp")

employee_list = []

for employee in employee_data:
    employee_dict = {}
    name = employee.find("PersonNm")
    employee_dict["name"] = name.text
    
    title = employee.find("TitleTxt")
    employee_dict["title"] = title.text
    
    compensation = employee.find("CompensationAmt")
    employee_dict["compensation"] = compensation.text
    
    employee_list.append(employee_dict)
```

```python
zuck_employee_df = pd.DataFrame(employee_list)
zuck_employee_df.to_csv('chan_zuckerberg_initiative_foundation2021_employee.csv')
zuck_employee_df
```

```python
recipient_data = zuck_data.find_all("GrantOrContributionPdDurYrGrp")
recipient_data
```

```python
recipient_list = []

for recipient in recipient_data:
    recipient_dict = {}
    recipient_name = recipient.find("BusinessNameLine1Txt")
    recipient_dict["business_name"] = recipient_name.text
    
    recipient_amount = recipient.find("Amt")
    recipient_dict["amount_received"] = recipient_amount.text
    
    recipient_purpose = recipient.find("GrantOrContributionPurposeTxt")
    recipient_dict["purpose"] = recipient_purpose.text

    recipient_city = recipient.find("CityNm")
    recipient_dict["city"] = recipient_city.text
    
    recipient_state = recipient.find("StateAbbreviationCd")
    recipient_dict["state"] = recipient_state.text
    recipient_list.append(recipient_dict)

    print(recipient_list)
    print("******")
```

```python
zuck_recipient_df = pd.DataFrame(recipient_list)
zuck_recipient_df.to_csv('chan_zuckerberg_initiative_foundation2021_recipient.csv')
zuck_recipient_df.head()
```

```python
contributor_data = zuck_data.find_all("ContributorInformationGrp")

contributor_list = []

for contributor in contributor_data:
    contributor_dict = {}
    name = contributor.find("BusinessNameLine1Txt")
    contributor_dict["contributor_name"] = name.text
    
    amount_contributed = contributor.find("TotalContributionsAmt")
    contributor_dict["contributed_amount"] = amount_contributed.text
    
    contributor_list.append(contributor_dict)
    
    print(contributor_dict)
```

```python
musk_contributor_df = pd.DataFrame(contributor_list)
musk_contributor_df
```
