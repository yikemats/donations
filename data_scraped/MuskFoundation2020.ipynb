```python
import requests
musk_foundation = requests.get("https://projects.propublica.org/nonprofits/download-xml?object_id=202143199349110414")
musk_foundation
```

```python
# The text is the raw xml of Form 990 tax document,obtained from ProPublica's non-profit explorer.
musk_xml = musk_foundation.text
print(musk_xml)
```

```python
from bs4 import BeautifulSoup
```

```python
musk_data = BeautifulSoup(musk_xml, "xml")
type(musk_data)
```

```python
musk_officer_data = musk_data.select("OfficerDirTrstKeyEmplGrp")
musk_officer_data
```

```python
import pandas as pd
```

```python
officer_data = musk_data.find_all("OfficerDirTrstKeyEmplGrp")

officer_list = []

for officer in officer_data:
    officer_dict = {}
    name = officer.find("PersonNm")
    officer_dict["name"] = name.text
    
    title = officer.find("TitleTxt")
    officer_dict["title"] = title.text
    
    officer_list.append(officer_dict)
```

```python
officer_list
```

```python
musk_officer_df = pd.DataFrame(officer_list)
musk_officer_df.to_csv('musk_foundation2020_officer.csv')
musk_officer_df
```

```python
recipient_data = musk_data.find_all("GrantOrContributionPdDurYrGrp")
recipient_data
```

```python
recipient_list = []

for recipient in recipient_data:
    recipient_dict = {}
    recipient_name = recipient.find("BusinessNameLine1Txt")
    recipient_dict["business_name"] = recipient_name.text
    
    recipient_amount = recipient.find("Amt")
    recipient_dict["amount_received"] = recipient_amount.text
    
    recipient_purpose = recipient.find("GrantOrContributionPurposeTxt")
    recipient_dict["purpose"] = recipient_purpose.text

    recipient_city = recipient.find("CityNm")
    recipient_dict["city"] = recipient_city.text
    
    recipient_state = recipient.find("StateAbbreviationCd")
    recipient_dict["state"] = recipient_state.text
    recipient_list.append(recipient_dict)

    print(recipient_list)
    print("******")
```

```python
musk_recipient_df = pd.DataFrame(recipient_list)
musk_recipient_df.to_csv('musk_foundation2020_recipient.csv')
musk_recipient_df
```

```python
contributor_data = musk_data.find_all("ContributorInformationGrp")

contributor_list = []

for contributor in contributor_data:
    contributor_dict = {}
    name = contributor.find("ContributorPersonNm")
    contributor_dict["contributor_name"] = name.text
    
    amount_contributed = contributor.find("TotalContributionsAmt")
    contributor_dict["contributed_amount"] = amount_contributed.text
    
    contributor_list.append(contributor_dict)
    
    print(contributor_dict)
```

```python
musk_contributor_df = pd.DataFrame(contributor_list)
musk_contributor_df
```
