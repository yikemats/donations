```python
import requests
zuck_biohub = requests.get("https://projects.propublica.org/nonprofits/download-xml?object_id=202203189349300800")
zuck_biohub
```

```python
# The text is the raw xml of Form 990 tax document,obtained from ProPublica's non-profit explorer.
zuck_xml = zuck_biohub.text
print(zuck_xml)
```

```python
from bs4 import BeautifulSoup
```

```python
zuck_data = BeautifulSoup(zuck_xml, "xml")
type(zuck_data)
```

```python
zuck_officer_data = zuck_data.select("Form990PartVIISectionAGrp")
zuck_officer_data
```

```python
import pandas as pd
```

```python
officer_list = []

for officer in zuck_officer_data:
    officer_dict = {}
    name = officer.find("PersonNm")
    officer_dict["name"] = name.text
    
    title = officer.find("TitleTxt")
    officer_dict["title"] = title.text
    
    compensation = officer.find("ReportableCompFromOrgAmt")
    officer_dict["reportable_compensation"] = compensation.text
    
    other_compensation = officer.find("OtherCompensationAmt")
    officer_dict["other_compensation"] = other_compensation.text
    
    officer_list.append(officer_dict)
```

```python
officer_list
```

```python
zuck_officer_df = pd.DataFrame(officer_list)
zuck_officer_df.to_csv('chan_zuckerberg_biohub2021_officer.csv')
zuck_officer_df
```

```python
recipient_data = zuck_data.find_all("RecipientTable")
recipient_data
```

```python
recipient_list = []

for recipient in recipient_data:
    recipient_dict = {}
    recipient_name = recipient.find("RecipientBusinessName")
    recipient_dict["business_name"] = recipient_name.text
    
    recipient_amount = recipient.find("CashGrantAmt")
    recipient_dict["amount_received"] = recipient_amount.text
    
    recipient_purpose = recipient.find("PurposeOfGrantTxt")
    recipient_dict["purpose"] = recipient_purpose.text

    recipient_city = recipient.find("CityNm")
    recipient_dict["city"] = recipient_city.text
    
    recipient_state = recipient.find("StateAbbreviationCd")
    recipient_dict["state"] = recipient_state.text
    recipient_list.append(recipient_dict)

    print(recipient_list)
    print("******")
```

```python
zuck_recipient_df = pd.DataFrame(recipient_list)
zuck_recipient_df["business_name"] = zuck_recipient_df["business_name"].str.strip("\n")
zuck_recipient_df.to_csv('chan_zuckerberg_biohub2021_recipient.csv')
zuck_recipient_df
```

```python
contributor_data = zuck_data.find_all("ContributorInformationGrp")

contributor_list = []

for contributor in contributor_data:
    contributor_dict = {}
    name = contributor.find("ContributorNum")
    contributor_dict["contributor_name"] = name.text
    
    business_contributer = contributor.find("BusinessNameLine1") 
    contributor_dict["business_contributor"] = business_contributer.text
    
    amount_contributed = contributor.find("TotalContributionsAmt")
    contributor_dict["contributed_amount"] = amount_contributed.text
    
    contributor_list.append(contributor_dict)
    
    print(contributor_dict)
```

```python
musk_contributor_df = pd.DataFrame(contributor_list)
musk_contributor_df
```
