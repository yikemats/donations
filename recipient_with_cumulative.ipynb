```python
import pandas as pd
```

```python
musk_data = pd.read_csv('musk_combined_recipient.csv')
musk_data.head()
```

```python
musk_data_df = pd.DataFrame(musk_data)
musk_data_df['musk_cumulative'] = musk_data_df['amount_received'].cumsum()
musk_data_df.head()
```

```python
musk_data_df.to_csv('musk_cumulative.csv')
```

```python
zuck_data = pd.read_csv('zuck_combined_recipient.csv')
zuck_data.head()
```

```python
zuck_data_df = pd.DataFrame(zuck_data)
zuck_data_df['zuck_cumulative'] = zuck_data_df['amount_received'].cumsum()
zuck_data_df.head()
```

```python
zuck_data_df.to_csv('zuck_cumulative.csv')
```

```python
musk_zuck_combined_df = pd.concat(map(pd.read_csv, ['musk_cumulative.csv', 'zuck_cumulative.csv']), ignore_index=False)
musk_zuck_combined_df
```

```python
musk_zuck_combined_df = musk_zuck_combined_df.sort_values(by=['timing'])
musk_zuck_combined_df = musk_zuck_combined_df[['timing', 'musk_cumulative', 'zuck_cumulative']]
musk_zuck_combined_df.head()
```

```python
musk_zuck_combined_df.to_csv('musk_zuck_combined.csv')
```
